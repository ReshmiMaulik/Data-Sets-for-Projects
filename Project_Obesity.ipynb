```
def map_gender(column):
    mapped=[]

    for value in column:

      if value=='Female':
        mapped.append(1)
      else: 
        mapped.append(0)

    return mapped
dataframe['gender_norms']=map_gender(dataframe['Gender'])
```

```
dataframe
```

```
def map_family_history_with_overweight(column):
    mapped=[]

    for value in column:

      if value=='yes':
        mapped.append(1)
      else: 
        mapped.append(0)

    return mapped
dataframe['family_history_with_overweight_norms']=map_family_history_with_overweight(dataframe['family_history_with_overweight'])
```

```
def map_FAVC(column):
    mapped=[]

    for value in column:

      if value=='yes':
        mapped.append(1)
      else: 
        mapped.append(0)

    return mapped
dataframe['FAVC_norms']=map_FAVC(dataframe['FAVC'])
```

```
def map_SMOKE(column):
    mapped=[]

    for value in column:

      if value=='yes':
        mapped.append(1)
      else: 
        mapped.append(0)

    return mapped
dataframe['SMOKE_norms']=map_SMOKE(dataframe['SMOKE'])
```

```
def map_SCC(column):
    mapped=[]

    for value in column:

      if value=='yes':
        mapped.append(1)
      else: 
        mapped.append(0)

    return mapped
dataframe['SCC_norms']=map_SCC(dataframe['SCC'])
```

```
def map_MTRANS(column):
    mapped=[]

    for value in column:

      if value=='Public_Transportation':
        mapped.append(1)
      else: 
        mapped.append(0)

    return mapped
dataframe['MTRANS_norms']=map_MTRANS(dataframe['MTRANS'])
```

```
dataframe
```

```
def map_NObeyesdad(column):
    mapped=[]

    for value in column:

      if value=='Overweight_Level_I':
        mapped.append(1)
      elif value=='Overweight_Level_II':
        mapped.append(2) 
      elif value=='Obesity_Type_III':
        mapped.append(3)   
      else: 
        mapped.append(0)

    return mapped
dataframe['NObeyesdad_norms']=map_NObeyesdad(dataframe['NObeyesdad'])
```

```
from sklearn.neighbors import KNeighborsClassifier
```
